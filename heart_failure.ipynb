```python
#import libraries
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from scipy import stats
```

```python
#show dataset
df= pd.read_csv(r'heart_failure_clinical_records_dataset.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df_copy=df.copy()
```

```python
y= df_copy['DEATH_EVENT']
x= df_copy.drop(columns='DEATH_EVENT',axis=1)
```

```python
# remove outlyers using z_score
from scipy.stats import zscore
z = np.abs(zscore(x))
mask = (z < 3).all(axis=1)
X = x.loc[mask]
y = y.loc[mask]
print(X.shape, y.shape)
```

```python
from sklearn.preprocessing import PowerTransformer
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
pt = PowerTransformer()
X= pt.fit_transform(X)
#x_scaled = scaler.fit_transform(x)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=0.95)
X_pca = pca.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_pca, y, test_size=0.2, random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)  # max_iter   converge
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
feature_columns = ['age', 'anaemia', 'creatinine_phosphokinase', 'diabetes', 'ejection_fraction', 'high_blood_pressure', 'platelets', 'serum_creatinine', 'serum_sodium', 'sex', 'smoking', 'time']
```

```python
def predict_heart_failure(age, anaemia, cpk, diabetes, ef, hbp, platelets, creatinine, sodium, sex, smoking, time):
    # Map string inputs for binary features back to numerical values
    anaemia_numeric = 1 if anaemia == 'yes' else 0
    diabetes_numeric = 1 if diabetes == 'yes' else 0
    hbp_numeric = 1 if hbp == 'yes' else 0
    sex_numeric = 1 if sex == 'male' else 0
    smoking_numeric = 1 if smoking == 'yes' else 0


    input_data = pd.DataFrame([{
        'age': age,
        'anaemia': anaemia_numeric,
        'creatinine_phosphokinase': cpk,
        'diabetes': diabetes_numeric,
        'ejection_fraction': ef,
        'high_blood_pressure': hbp_numeric,
        'platelets': platelets,
        'serum_creatinine': creatinine,
        'serum_sodium': sodium,
        'sex': sex_numeric,
        'smoking': smoking_numeric,
        'time': time
    }])
    input_data = input_data[feature_columns]
    input_power = pt.transform(input_data)
    input_pca = pca.transform(input_power)
    prediction = model.predict(input_pca)[0]
    prediction_proba = model.predict_proba(input_pca)[0]

    result = f"Prediction: {'Death' if prediction == 1 else 'No Death'}"
    if prediction == 1:
        result += f" (Probability of Death: {prediction_proba[1]:.2f})"
    else:
        result += f" (Probability of No Death: {prediction_proba[0]:.2f})"

    return result
```

```python
import gradio as gr

demo = gr.Interface(
    fn=predict_heart_failure,
    inputs=[
        gr.Number(label="Age"),
        gr.Radio(['no', 'yes'], label="Anaemia"),
        gr.Number(label="CPK"),
        gr.Radio(['no', 'yes'], label="Diabetes"),
        gr.Number(label="Ejection Fraction"),
        gr.Radio(['no', 'yes'], label="High Blood Pressure"),
        gr.Number(label="Platelets"),
        gr.Number(label="Serum Creatinine"),
        gr.Number(label="Serum Sodium"),
        gr.Radio(['female', 'male'], label="Sex"),
        gr.Radio(['no', 'yes'], label="Smoking"),
        gr.Number(label=" Time")
    ],
    outputs="text"
)

demo.launch()
```
